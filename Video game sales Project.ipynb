Remove Unwanted Column

```python
if 'Unnamed: 0' in game_data.columns:
    game_data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Handling missing values 
game_data['Team'] = game_data['Team'].fillna('Unknown')
game_data['Summary'] = game_data['Summary'].fillna('No summary available')
game_data['Rating'] = game_data['Rating'].fillna(game_data['Rating'].mean())

```

```python
# Convert release date to datetime format
game_data['Release Date'] = pd.to_datetime(game_data['Release Date'], errors='coerce')
game_data = game_data.dropna(subset=['Release Date'])
```

```python
game_data.duplicated().sum()
game_data.drop_duplicates(inplace=True)
```

```python
def clean_title(title):
    return title.lower().replace("-","").replace("(","").replace(")","").replace(":","").replace("&","").replace(",","").replace("[","").replace("]","").replace("'","").replace("'","").replace(" " ","").replace(" " ","")
```

```python
# Data Cleaning
game_data['Title'] = game_data['Title'].astype(str).apply(clean_title)
game_data['Team'] = game_data['Team'].astype(str).apply(clean_title)
game_data['Genres'] = game_data['Genres'].astype(str).apply(clean_title)
game_data['Summary'] = game_data['Summary'].astype(str).apply(clean_title)
game_data['Reviews'] = game_data['Reviews'].astype(str).apply(clean_title)
game_data['Rating'] = game_data['Rating'].astype(str).apply(clean_title)
game_data['Playing'] = game_data['Playing'].astype(str).apply(clean_title)
```

```python
# list of column that having 'k' value
k_col = ['Times Listed', 'Number of Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist']

# converting 'k' value to numeric
for col in k_col:
    if col in game_data.columns:
        game_data[col] = game_data[col].astype(str).str.upper()
        game_data[col] = game_data[col].apply(lambda x: float(x.replace('K', '')) * 1000 if 'K' in x else x)
        game_data[col] = pd.to_numeric(game_data[col], errors='coerce')
```

```python
game_data.to_csv('Clean_game_data.csv', index=False)
```

```python
# ==========  Load Datasets ==========
game_data = pd.read_csv("Clean_game_data.csv")
sales_data = pd.read_csv("Clean_sales_data.csv")
merged_data = pd.read_csv("merged_data.csv")


# ==========  INSERT GAME_DATA ==========
print("Inserting Game_data...")
for index, row in game_data.iterrows():
    cursor.execute("""
        INSERT INTO Game_data (Title, Release_Date, Team, Rating, Times_Listed, 
                               Number_of_Reviews, Genres, Summary, Reviews, 
                               Playing, Plays, Backlogs, Wishlist)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Title'], row['Release Date'], str(row['Team']), row['Rating'],
        row['Times Listed'], row['Number of Reviews'], str(row['Genres']),
        str(row['Summary']), str(row['Reviews']), 
        row['Playing'], row['Plays'], row['Backlogs'], row['Wishlist']
    ))

# ==========  INSERT VGSALES ==========
print("Inserting VGSALES...")
for index, row in sales_data.iterrows():
    cursor.execute("""
        INSERT INTO VGSALES (Name, Platform, Year, Genre, Publisher, 
                             NA_Sales, EU_Sales, JP_Sales, Other_Sales, Global_Sales)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Name'], row['Platform'], str(row['Year']), row['Genre'], row['Publisher'],
        row['NA_Sales'], row['EU_Sales'], row['JP_Sales'], 
        row['Other_Sales'], row['Global_Sales']
    ))

# ==========  INSERT MERGED_DATA ==========
print("Inserting Merged_data...")
for index, row in merged_data.iterrows():
    cursor.execute("""
        INSERT INTO merged_data (Title, Release_Date, Team, Rating, Times_Listed, 
                                 Number_of_Reviews, Genres, Summary, Reviews, 
                                 Playing, Plays, Backlogs, Wishlist, Platform, Year,
                                 Genre, Publisher, NA_Sales, EU_Sales, JP_Sales,
                                 Other_Sales, Global_Sales)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Title'], row['Release Date'], str(row['Team']), row['Rating'], row['Times Listed'],
        row['Number of Reviews'], str(row['Genres']), str(row['Summary']), str(row['Reviews']),
        row['Playing'], row['Plays'], row['Backlogs'], row['Wishlist'], 
        row['Platform'], row['Year'], row['Genre'], row['Publisher'],
        row['NA_Sales'], row['EU_Sales'], row['JP_Sales'], row['Other_Sales'], row['Global_Sales']
    ))

# ========== 7. COMMIT ==========
conn.commit()
print("✅ All data inserted successfully!")

cursor.close()
conn.close()
```
